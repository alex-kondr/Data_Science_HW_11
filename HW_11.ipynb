```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from keras.datasets import imdb
from keras.utils import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense, LSTM, GRU, TimeDistributed, Bidirectional
```

```python
max_features = 10000
```

```python
(train_data, y_train), (test_data, y_test) = imdb.load_data(num_words=max_features)
```

```python
def RNN(train_data, y_train, test_data, maxlen: int, units: int, batch_size: int = 128, network = SimpleRNN) -> dict:
    train_data = pad_sequences(train_data, maxlen=maxlen)
    test_data = pad_sequences(test_data, maxlen=maxlen)
    model = Sequential()
    model.add(Embedding(max_features, 16, input_length=maxlen))
    model.add(network(units))
    model.add(Dense(1, activation="sigmoid"))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    history = model.fit(train_data, y_train, epochs=5, batch_size=200, validation_split=0.2)

    return history
```

```python
def plot_history(history: dict):
    plt.figure(figsize=(15, 6))
    plt.plot(history.history["loss"], label="loss")
    plt.plot(history.history["accuracy"], label="accuracy")
    plt.plot(history.history["val_loss"], label="val_loss")
    plt.plot(history.history["val_accuracy"], label="val_accuracy")
    plt.legend()
    plt.title("Accuracy-Epoch")
    plt.xlabel("Epoch")
    plt.grid()
```

```python
history = RNN(train_data, y_train, test_data, 500, 16, 128)
plot_history(history)
```

```python
history = RNN(train_data, y_train, test_data, 500, 16, 64)
plot_history(history)
```

```python
history = RNN(train_data, y_train, test_data, 300, 16, 128)
plot_history(history)
```
